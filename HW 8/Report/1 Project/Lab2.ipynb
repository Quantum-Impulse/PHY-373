```python
import numpy as np
import matplotlib.pyplot as plt

# Example frequency sweep
f = np.logspace(1, 4.3, 500)  # Frequencies from ~10 Hz to ~20 kHz
# Hypothetical transfer functions for each knob setting:
# These would be replaced by measured data or a more complex filter model.
# For demonstration, let's assume simple filter responses:
# Bass: low-pass characteristic
R_bass = 10e3
C_bass = 47e-9
H_bass = 1 / np.sqrt(1 + (2*np.pi*f*R_bass*C_bass)**2)

# Mid: band-pass characteristic (using a simple approximation)
# Center around 1kHz, let's fake a response:
f0_mid = 1100  # Hz from our calc
Q_mid = 1.0    # arbitrary quality factor
H_mid = (f/f0_mid)*Q_mid / np.sqrt(1 + ((f/f0_mid)*Q_mid)**2)

# Treble: high-pass characteristic
R_treb = 3.6e3
C_treb = 4.7e-9
H_treb = (2*np.pi*f*R_treb*C_treb)/np.sqrt(1+(2*np.pi*f*R_treb*C_treb)**2)

plt.figure(figsize=(8,4))
plt.semilogx(f, 20*np.log10(H_bass), label='Bass Control')
plt.semilogx(f, 20*np.log10(H_mid),  label='Mid Control')
plt.semilogx(f, 20*np.log10(H_treb), label='Treble Control')
plt.grid(True, which='both', ls='--')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude (dB)')
plt.title('Approximate Frequency Response of EQ Sections')
plt.legend()
plt.savefig('eq_frequency_response.png', dpi=300)
plt.show()
```
